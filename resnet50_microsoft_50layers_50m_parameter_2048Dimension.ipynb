```python
train_dir = "/content/drive/MyDrive/TT Library/AI Model/Data/Pneumonia/train"
test_dir = "/content/drive/MyDrive/TT Library/AI Model/Data/Pneumonia/test"
```

```python
X_train = []
Y_train = []
```

```python
for i in os.listdir(train_dir):
  for j in os.listdir(train_dir+"/"+i):
    img = cv2.imread(train_dir+"/"+i+"/"+j)
    img = cv2.resize(img,(150,150))
    img = image.img_to_array(img)
    img = resnet50.preprocess_input(np.expand_dims(img.copy(),axis=0))
    img = model.predict(img)
    img = img.flatten()
    X_train.append(img)
    Y_train.append(i)
```

```python
X_test = []
Y_test = []
for i in os.listdir(test_dir):
  for j in os.listdir(test_dir+"/"+i):
    img = cv2.imread(test_dir+"/"+i+"/"+j)
    img = cv2.resize(img,(150,150))
    img = image.img_to_array(img)
    img = resnet50.preprocess_input(np.expand_dims(img.copy(),axis=0))
    img = model.predict(img)
    img = img.flatten()
    X_test.append(img)
    Y_test.append(i)
```

```python
from sklearn import linear_model
from sklearn.metrics import accuracy_score,confusion_matrix
model = linear_model.LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)
pred = model.predict(X_test)
print(accuracy_score(Y_test,pred))
print(confusion_matrix(Y_test,pred))
```

```python
from sklearn import ensemble
model = ensemble.RandomForestClassifier()
model.fit(X_train,Y_train)
pred = model.predict(X_test)
print(accuracy_score(Y_test,pred))
print(confusion_matrix(Y_test,pred))
```
